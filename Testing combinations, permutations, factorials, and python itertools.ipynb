```python
def nCr(n, r):
    return fac(n) / (fac(r) * fac(n - r))
```

```python
nCr(10,2)
```

```python
def p(n, r):
    print(n, r)
    if r > 0:
        print(n - r)
        print('recurse')
        return (n - r - 1) * p(n, r - 1)
    else:
        print(n - r)
        print('escape')
        return n
    
p(7,2)
```
